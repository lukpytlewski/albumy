```python
import pandas as pd
import urllib.request
import ssl
def main():
    ssl._create_default_https_context = ssl._create_unverified_context
    r = urllib.request.urlopen('https://www.officialcharts.com/chart-news/the-best-selling-albums-of-all-time-on-the-official-uk-chart__15551/')
    print(r.status)
    print(r)
main()
```

```python
data = pd.read_html('https://www.officialcharts.com/chart-news/the-best-selling-albums-of-all-time-on-the-official-uk-chart__15551/', header=0)
```

```python
df=pd.DataFrame(data[0])
df.head()
```

# Zamień nagłówki kolumn na polskie odpowiedniki: ['TYTUŁ','ARTYSTA','ROK','MAX POZ']

```python
df.rename(columns={'POS':'POZYCJA','TITLE':'TYTUŁ','ARTIST':'ARTYSTA', 'YEAR':'ROK', 'HIGH POSN':'POZ MAX'}, inplace=True)
df.head()
```

# Ilu pojedynczych artystów znajduje się na liście?

```python
df['ARTYSTA'].nunique()
```

# Które zespoły pojawiają się najczęściej na liście?

```python
df['ARTYSTA'].value_counts()[:5]
```

# Zmień nagłówki kolumn, tak aby każdy z nich rozpoczynał się od wielkiej litery, a pozostałe były wprowadzone małymi literami.

```python
df.rename(str.capitalize, axis='columns', inplace=True)
df.head()
```

# Wyrzuć z tabeli kolumnę ‘Max Poz’.

```python
df.drop('Poz max',axis=1, inplace=True)
df.head()
```

# W którym roku wyszło najwięcej albumów znajdujących się na liście?

```python
df['Rok'].value_counts()[:5]
```

# Ile albumów wydanych między 1960 a 1990 rokiem włącznie znajduje się na liście?

```python
df[(df['Rok']>=1960) & (df['Rok']<=1990)]
```

```python
liczba_albumow=len(df[(df['Rok']>=1960) & (df['Rok']<=1990)])
liczba_albumow
```

# W którym roku wydany został najmłodszy album na liście?

```python
df.sort_values(by='Rok',ascending=False)[:5]
```

# Przygotuj listę najwcześniej wydanych albumów każdego artysty, który znalazł się na liście.

```python
df.sort_values(by=['Artysta','Rok'], inplace=True)
df.head()
```

```python
df2=df.groupby(pd.Grouper(key='Artysta')).agg({'Tytuł':'first','Rok':'min'})
df2.head()
```
